```python
import numpy as np

N = 1000000
M = 3
heads = np.zeros(M+1)
for i in range(N):
    flips = np.random.randint(0, 2, M)
    h, _ = np.bincount(flips, minlength=2)
    heads[h] += 1
prob = heads / N
print("Probabilities: ", np.array2string(prob))
```

```python
import numpy as np

match = 0
for i in range(1000000):
    a = np.random.randint(0,364)
    b = np.random.randint(0,364)
    if a == b:
        match += 1
print("Probability of a random match: ", match/1000000)
```

```python
list_matches = []
for m in range(2,32):
    matches = 0
    for n in range(100000):
        match = 0
        b = np.random.randint(0,364,m)
        for i in range(m):
            for j in range(m):
                if (i != j) and (b[i] == b[j]):
                    match += 1
        if match != 0:
            matches += 1
    prob_matches = matches/100000
    list_matches.append(prob_matches)
    print(m, prob_matches)
```

```python
import matplotlib.pyplot as plt

people = list(range(2, 32))
probs = list_matches
plt.figure()
plt.plot(people, probs, marker='o')
plt.axhline(0.5, linestyle='--')
plt.xlabel('Probability of a shared birthday')
plt.ylabel('Number of people')
plt.show()
```
